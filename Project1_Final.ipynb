## 2019년 남자/여자 인구 현황

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15 
matplotlib.rcParams['axes.unicode_minus'] = False 
```

### 1. 2019년 남자 인구

```python
df_m= pd.read_csv('C:/project/manPop_2019.csv')
df_m = df_m.iloc[:, 0:13]
df_m.head()
```

```python
df_m.columns = df_m.columns.str.replace('_남자 인구수', '')
df_m.columns = df_m.columns.str.replace('2019년', '')
df_m.columns
```

```python
df_m.head()
```

```python
# 2019년 1월 & 12월 남자 인구수 변화

index = np.arange(17)
w = 0.25
plt.bar(index, df_m['1월'], width = w, color = '#ED7464', label = '1월')
plt.bar(index + w, df_m['12월'], width = w, color = '#611487', label = '12월')

plt.title('2019년 남자 인구수 변화', fontsize = 15, pad = 10)
plt.ticklabel_format(axis = 'y', useOffset = False, style = 'plain')
plt.legend()
plt.xlabel('시/도', fontsize = 10)
plt.ylabel('인구수 (명)', fontsize = 10)
plt.xticks(index + w, df_m['행정구역'], rotation = 60, fontsize = 8)
plt.yticks(fontsize = 10)

plt.show()
```

```python
# 2019년 월별 남자 총인구수 변동

total_m = []
for month in range(1,13):
    total_m.append(df_m['{}월'.format(month)].sum())

mth = []
for i in range(1,13):
    mth.append(str(i) + "월")
    
plt.plot(mth, total_m, color = '#f00c85', marker = 'o', mec = 'w', mew = 2, lw = 2)
plt.title('2019년 남자 월별 총인구수 변동', fontsize = 15, pad = 10)
plt.ticklabel_format(axis = 'y', useOffset = False, style = 'plain')
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)

plt.show()
```

```python
# 2019년 남자 100만명 이상 지역

m_m = df_m[df_m['12월'] >= 1000000][['행정구역', '12월']]

plt.bar(m_m['행정구역'], m_m['12월'], color = '#c2ab95')
plt.title('2019년 12월 남자 인구 100만명 이상 지역', fontsize = 15, pad = 10)
plt.ticklabel_format(axis = 'y', useOffset = False, style = 'plain')
plt.xticks(fontsize = 10, rotation = 45)
plt.yticks(fontsize = 10)

plt.show()
```

```python
# 2019년 지역별 남자 인구 비율 (백만명 이상)

matplotlib.rcParams['font.size'] = 9 

plt.figure(figsize = (4,4))
ratio = m_m['12월']
labels = m_m['행정구역']
explode = [0.05] * 8

plt.pie(ratio, labels = labels, labeldistance = 1.07, autopct = '%.1f%%', explode = explode)
plt.title('2019년 12월 지역별 남자 인구 비율 (100만명 이상)', fontsize = 12, pad = 10)



plt.show()
```
